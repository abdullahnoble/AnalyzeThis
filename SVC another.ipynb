```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns',25)
#from sklearn.model_selection import ShuffleSplit
```

```python
df = pd.read_csv('development_dataset.csv')
df.describe()
```

```python
#converting object type VAR14 to float type
df['VAR14'].replace(to_replace='.',value=np.nan,inplace=True)
df['VAR14']=pd.to_numeric(df['VAR14'])
df.dtypes
```

```python
# mapping assigned lines
df['VAR21']=df['VAR21'].map({'Low':1,'Medium':2,'High':3})
```

```python
for col in df.columns:
    df[col]=df[col].fillna(df[col].median())
```

```python
#line = df['VAR21']
df2 = df.drop(labels=['VAR1','VAR19','VAR14','VAR10','VAR18'],axis=1)
df2.head()
```

```python
for cols in df2.columns:
    if (cols!='VAR21'):
        df2 = df2[np.abs(df2[cols]-df2[cols].mean()) <= (3*df2[cols].std())]
print(df2.shape)
```

```python
line = df2['VAR21']
train = df2.drop(labels=['VAR21'],axis=1)
train.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(train,line,test_size=0.2)  
print(xtrain.shape)
print(xtest.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
trainx2 = StandardScaler().fit(xtrain).transform(xtrain.astype(float))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
svc = SVC()
```

```python
svc.fit(trainx2,ytrain)
```

```python
ans = svc.predict(xtest)
```

```python
accuracy_score(ytest,ans)
```
