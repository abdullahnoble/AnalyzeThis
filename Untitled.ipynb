```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',25)
```

```python
df = pd.read_csv('development_dataset.csv')
```

```python
df.head()
```

```python
test = pd.read_csv('leaderboard_dataset.csv')
test.head()
```

```python
df['VAR14'].replace(to_replace='.',value=np.nan,inplace=True)
df['VAR14']=pd.to_numeric(df['VAR14'])
df.dtypes
```

```python
df['VAR21']=df['VAR21'].map({'Low':1,'Medium':2,'High':3})
```

```python
line = df['VAR21']
train = df.drop(labels = ['VAR21','VAR1'],axis=1)
test = test.drop(labels = ['VAR1'],axis=1)
train.tail()
```

```python
data = pd.DataFrame(np.ones(df.shape[0],dtype=bool),columns=['Train'])
data2 = pd.DataFrame(np.zeros(test.shape[0],dtype=bool),columns=['Train'])
```

```python
trainf = pd.concat([train,data],axis='columns')
testf = pd.concat([test,data2],axis='columns')
testf.tail()
```

```python
dfr = pd.concat((trainf,testf)).reset_index(drop=True)
dfr.head()
```

```python
for col in trainf:
    dfr[col]=dfr[col].fillna(trainf[col].median())
```

```python
dfr.head()
```

```python
trainx = dfr[dfr.Train==True]
trainx = trainx.drop(labels=['Train'],axis=1)
trainx.head()
```

```python
dfr.isnull().any()
line.value_counts()
```

```python
from sklearn import preprocessing
trainx.dtypes
```

```python
trainx2 = preprocessing.StandardScaler().fit(trainx).transform(trainx.astype(float))
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(trainx2,line,test_size=0.25)  
print(xtrain.shape)
print(xtest.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
krange=range(1,30)
scores={}
score_list = []
for k in krange:
    knn=KNeighborsClassifier(n_neighbors=k).fit(xtrain,ytrain)
    ypred=knn.predict(xtest)
    scores[k]=metrics.accuracy_score(ytest,ypred)
    score_list.append(metrics.accuracy_score(ytest,ypred))
    
print(score_list)
```

```python
plt.plot(krange,score_list)
plt.xlabel('value of k for knn')
```

```python
neigh = KNeighborsClassifier(n_neighbors = 30).fit(xtrain,ytrain)
```

```python
testx1 = dfr[dfr.Train==False]
testx1['VAR14'].replace(to_replace='.',value=np.nan,inplace=True)
testx1['VAR14']=pd.to_numeric(testx1['VAR14'])
testx1.dtypes
```

```python
testx1['VAR14']=testx1['VAR14'].fillna(trainf['VAR14'].median())
testx1.isnull().any()
```

```python
testx1= testx1.drop(labels=['Train'],axis=1)
testx1.head()
testx2 = preprocessing.StandardScaler().fit(testx1).transform(testx1.astype(float))
```

```python
ans = neigh.predict(testx2)
```
